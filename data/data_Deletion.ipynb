```python
# importing necessary libraries
import pandas as pd
```

```python
# loading the dataset
data = pd.read_csv('creditcard.csv')
data.shape
```

```python
# select 20% from each class for removal (stratified by Class)
fraud_indices = data[data['Class'] == 1].index
non_fraud_indices = data[data['Class'] == 0].index
num_fraud_to_remove = int(0.32 * len(fraud_indices))
num_non_fraud_to_remove = int(0.32 * len(non_fraud_indices))
fraud_indices_to_remove = fraud_indices[:num_fraud_to_remove]
non_fraud_indices_to_remove = non_fraud_indices[:num_non_fraud_to_remove]
indices_to_remove = fraud_indices_to_remove.union(non_fraud_indices_to_remove)
# remove the selected rows
data_dropped = data.drop(indices_to_remove) 
data_dropped.shape
```

```python
# save csv file 
data_dropped.to_csv('creditcard_sample.csv', index=False)
data_dropped.head()
```
